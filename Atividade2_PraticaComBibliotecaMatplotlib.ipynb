1. Importe as bibliotecas necessárias:
*   Matplotlib.pyplot como plt
*   NumPy como np





```python
import matplotlib.pyplot as plt
import numpy as np
```

2. Crie um conjunto de dados:
*   Crie um array NumPy chamado x com 100 valores, variando de 0 a 10.
*   Crie um array NumPy chamado y1 contendo os valores da função seno para cada valor de x.
*   Crie um array NumPy chamado y2 contendo os valores da função cosseno para cada valor de x.





```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

3. Crie um gráfico de linha:
*   Use a função plot() para criar um gráfico de linha com x no eixo x e y1 no eixo y.
*   Personalize o gráfico com um título, rótulos para os eixos x e y.



```python
plt.plot(x, y1, color='blue')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Gráfico de Linha')
plt.grid(True)
plt.show()
```

4. Crie um gráfico de dispersão:
*   Use a função scatter() para criar um gráfico de dispersão com x no eixo x e y2 no eixo y.
*   Personalize o gráfico com um título, rótulos para os eixos x e y.



```python
plt.scatter(x, y2, color='blue')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Gráfico de Dispersão')
plt.grid(True)
plt.show()
```

5. Crie um gráfico de barras:
*   Crie um novo conjunto de dados chamado valores com 5 valores inteiros de sua escolha.
*   Crie uma lista de rótulos para as barras correspondentes aos valores.
*   Use a função bar() para criar um gráfico de barras com valores e rótulos.
*   Personalize o gráfico com um título, rótulos para os eixos x e y.





```python
valores = [12, 15, 20, 6, 37]
rotulos = ['JAN', 'FEV', 'MAR', 'ABR', 'MAI']
barras = plt.bar(rotulos, valores, color = 'blue')
plt.xlabel('Meses')
plt.ylabel('Quantidades')
plt.title('Gráfico de Barras')
plt.grid(True)
plt.savefig('grafico_barras.png')
plt.show()
```

6. Crie um subplot:
*   Crie uma figura com dois subplots empilhados verticalmente.
*   No primeiro subplot, plote o gráfico de linha.
*   No segundo subplot, plote o gráfico de dispersão.
*   Personalize cada subplot com títulos adequados.

```python
# Criando os subplots:
fig, axs = plt.subplots(2, 1)

# Primeiro subplot:
axs[0].plot(x, y1, color = 'blue')
axs[0].set_title('Subplot 1: Gráfico de Linha')

# Segundo subplot:
axs[1].scatter(x, y2, color = 'blue')
axs[1].set_title('Subplot 2: Gráfico de Dispersão')

# Ajustando o layout:
plt.tight_layout()

# Exibindo os subplots:
plt.show()
```
