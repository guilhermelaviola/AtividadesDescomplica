1. Importe as bibliotecas necessárias:
*   Matplotlib.pyplot como plt
*   NumPy como np





```python
import matplotlib.pyplot as plt
import numpy as np
```

2. Crie um conjunto de dados:
*   Crie um array NumPy chamado x com 100 valores, variando de 0 a 10.
*   Crie um array NumPy chamado y1 contendo os valores da função seno para cada valor de x.
*   Crie um array NumPy chamado y2 contendo os valores da função cosseno para cada valor de x.





```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

3. Crie um gráfico de linha:
*   Use a função plot() para criar um gráfico de linha com x no eixo x e y1 no eixo y.
*   Personalize o gráfico com um título, rótulos para os eixos x e y.



```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, color='blue')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Gráfico de Linha')
plt.grid(True)
plt.show()
```

4. Crie um gráfico de dispersão:
*   Use a função scatter() para criar um gráfico de dispersão com x no eixo x e y2 no eixo y.
*   Personalize o gráfico com um título, rótulos para os eixos x e y.



```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y2, color='blue')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Gráfico de Dispersão')
plt.grid(True)
plt.show()
```
